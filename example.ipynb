# Extraction of Tables From Image

Documents could be very useful source of data, in this example we will introduce approach for automated extraction of tables from documents. Many packages focused on text mining from pdf files already exists, but functions which extract tables aren't always reliable. Our approach is based on computer vision and OCR. In first step algorithm identify location of tables in page, then OCR is used to read cells and record it into data structure which can be easily converted to standard data fomats like dictionary or DataFrame.

## Document

To be able to read tables we need to convert page of the document to image format. Picture bellow represents our example document page, as we can see it contains 4 different tables.

```python
from IPython.display import Image
Image("data/page_images/page_8.jpg")
```

```python
# this code has function that help us to spare space in this jupyter notebook by allowing us to display more dataframes side by side
from IPython.display import display_html
from itertools import chain,cycle
def display_side_by_side(*args,titles=cycle([''])):
    html_str=''
    for df,title in zip(args, chain(titles,cycle(['</br>'])) ):
        html_str+='<th style="text-align:center">'
        html_str+=df.to_html().replace('table','table style="display:inline"')
        html_str+='</th>'
    display_html(html_str,raw=True)
```

Our table reader can be used very easily, we just need path to image with document page and function `table_reader` will extract and return list with tables.

```python
from table_reader import table_reader

path = 'data/page_images/page_8.jpg'
tables = table_reader(path)
```

Tables can be converted to DataFrame using method `table_to_dataframe` or to dictionary with `table_to_dict`. As we can see bellow all 4 tables were extracted.

```python
display_side_by_side(tables[0].table_to_dataframe(True), tables[1].table_to_dataframe(True))
```

```python
display_side_by_side(tables[2].table_to_dataframe(True), tables[3].table_to_dataframe(True))
```
